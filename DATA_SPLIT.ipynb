```python
import glob
import pandas as pd
from sklearn.model_selection import train_test_split
import shutil
import os
```

```python
data_dir = 'forSplit'
file_format = 'jpg'

img_dir = ['/*.png', '/*.jpg' ]
label_format = '.txt'

seed = 1004

img_file_list = []
label_file_list = []
```

```python
for img in img_dir:
    temp = glob.glob(data_dir+img)
    for i in temp:
        img_file_list.append(i)
        temp_name = i.split('\\') # 이미지 파일 경로에서 \\ 기준으로 string 분리
        temp_name.reverse() # 역순으로 분리된 string들을 역순으로 배치하여 이미지 파일이름을 얻어냄
        extention = temp_name[0].split('.') # 확장자 이름 얻기
        extention = '.'+extention[1]
        label_name = temp_name[0].split(extention) # 이미지 파일의 확장자 이름을 때어낸 이름을 줌
        label_file_list.append(data_dir+'\\'+label_name[0]+label_format) # 라벨링 파일 이름에 확장자를 붙여줌

print(img_file_list)
print(label_file_list)
```

```python
print(len(img_file_list))
print(len(label_file_list))
```

```python
df = pd.DataFrame(
{
    'img_dir' : img_file_list,
    'label_dir': label_file_list
})
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df['img_dir'], 
                                                    df['label_dir'], 
                                                    test_size=0.3, 
                                                    shuffle=True, 
                                                    random_state=seed
                                                   )
```

```python
valid_x, test_x, valid_y, test_y = train_test_split(X_test, 
                                                    y_test, 
                                                    test_size=0.3, 
                                                    shuffle=True, 
                                                    random_state=seed
                                                   )

# 결과 -> training: test : valid = 70 : 21 : 9
```
